**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

**Data collection and Processing**

```python
# Loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
## Loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
## Adding the target column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

2 --> Malignant

```python
data_frame.groupby('label').mean()
```

```python
## Separating the features and target
X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting the data into Training and Testing data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state= 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

**Building the Nueral network**

```python
## Importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
## Setting up the layers of Nueral network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #Flatten converts input into single dimenaional array with no. of i/p features provided
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
## Compiling the nueral network
model.compile(optimizer= 'adam',
              loss = 'sparse_categorical_crossentropy', # classification problem
              metrics = ['accuracy'])
```

```python
## Training the nueral network

history = model.fit( X_train_std, Y_train, validation_split =0.1, epochs=10)


## For model we need to reduce the loss and improve the accuracy
```

**Visualizing the accuracy and loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.legend(['training data', 'validation data'], loc ='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.legend(['training data', 'validation data'], loc ='lower right')
```

 **Accuracy of the model on test data**

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print('Test Accuracy:', accuracy)
```
